Fonction pour étudier la complétude de la table en fonction des pays.
La fonction renvoie le pourcentage de case remplies par pays

```python

def completeness_per_country(df):
    missing_values_per_country = defaultdict(lambda: {"missing": 0, "N_case": 0})
    for i in range(len(df)):
        row = df.iloc[i]
        cc = df["country_code"][i]
        missing_case = row.isnull().sum() 
        N_case_per_country = row.count()
        missing_values_per_country[cc]["missing"] += missing_case
        missing_values_per_country[cc]["N_case"] += missing_case + N_case_per_country
    
    print("Complétude par pays:")
    for country, case in missing_values_per_country.items():
        missing = case["missing"]
        N_case_tot = case["N_case"]
        print(f"Pays : {country}, Pourcentages de complétudes : {(missing / N_case_tot) * 100}%")
    
    return missing_values_per_country
```

Fonction pour regarder si les données ont bien été remplie tous les jours.
La fonction renvoie un pourcentage de jour remplis par pays
On utilise une fonction auxiliaire check_value pour regarder si les valeur sont bien des dates et surtout pour ajouter chaque date à un dictionnaire

```python
def check_value(value):
    unique_dates = set()
    try:
        date = datetime.strptime(value, "%d/%m/%Y") ## converti en {date}
        unique_dates.add(date.strftime("%d/%m/%Y")) ## converti en str
    except ValueError:
        print("Date invalide")

    return unique_dates ## retourne la liste des dates sans doublon

def timeliness_per_country(df):
    dates_by_month_per_country = defaultdict(set)   
    for i in range(len(df)):
        value = df["date_day"][i]
        cc = df["country_code"][i]
        valid_dates = check_value(value)

        for date in valid_dates:
            date_obj = datetime.strptime(date, "%d/%m/%Y")
            dates_by_month_per_country[(cc, date_obj.year, date_obj.month)].add(date_obj.day)
    
    for (cc, year, month), days in sorted(dates_by_month_per_country.items()):
        if month in [4, 6, 9, 11]: 
            N_days = 30
        elif month == 2:  
            N_days = 28
        else:
            N_days = 31
        
        percent = (len(days) / N_days) * 100
        print(f"Pays : {cc}, Année : {year}, Mois : {month}, Jours où les données sont remplies: {len(days)}/{N_days}")
        print(f"Pourcentage de jour où les données ont été remplies pour le mois: {percent:.2f}%\n")
```

Fonction pour étudier l'intégrité des données 
La fonction compare le nombre de session entre la partie tracking et reporting et renvoie le pourcentage d'écart (en V.A) entre les deux
Initilament nous voulions comparer le nombre de session entre 09/2024 et 12/2024 
mais nous n'avons que les résultats d'octobre 2024 pour le fichier tracking (grosse volumétrie de donnée) nous étudierons l'écart que sur 10/2024
l'écart est signicatif mais il faut prendre en compte les contraintes matériel et les ressources à notre disposition

```python
def period_filter(df, date_column):
    filtered_date = df[(df[date_column] >= datetime(2024,9,1)) & (df[date_column] <= datetime(2024,12,1))]
    return filtered_date




def integrity_df(f1,f2):
    df1 = load_file(f1)
    df2 = load_file_track(f2)
    df1["date_day"] = pd.to_datetime(df1["date_day"], format="%d/%m/%Y", errors='coerce')
    df2["date"] = pd.to_datetime(df2["date"])
    df2_summed = df2.groupby('date')['session_id'].size().reset_index(name='sessions_count')
    result = df1.merge(df2_summed, left_on='date_day', right_on='date', how='left')
    result = result.drop("date", axis=1)
    return result

def integrity_per_month(df_merge, date_column):
    print("intégrité pour le pays FR")
    df_filtered = period_filter(df_merge, date_column)
    integrity_per_m = defaultdict(dict)

    for _, row in df_filtered.iterrows():
        year = row[date_column].year
        month = row[date_column].month

        # Calculer la différence entre session_count et session_sum pour chaque ligne
        session_diff = abs(row["sessions_count"] - row["number_of_sessions"])
        if (year, month) not in integrity_per_m:
            integrity_per_m[(year, month)] = {"session_diff": 0}

        integrity_per_m[(year, month)]["session_diff"] += session_diff

    # Convertir les résultats en DataFrame
    result_df = pd.DataFrame([
        {"year": year, "month": month, "integrity_result": data["session_diff"]}
        for (year, month), data in integrity_per_m.items()
    ])

    # Afficher les résultats
    print("pourcentage d'écart entre le reporting et le tracking & Résultats d'intégrité mensuelle :")
    print(result_df)

    return result_df
```
